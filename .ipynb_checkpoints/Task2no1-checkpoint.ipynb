```python
import sklearn.datasets
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import fetch_olivetti_faces
```

```python
olivetti_faces = fetch_olivetti_faces()
```

```python
faces = olivetti_faces
```

```python
Xall = faces.data.T
print(Xall.shape)
```

```python
print(Xall.min(), Xall.max())
```

```python
Xsub = Xall[:,0:200]
print(Xsub.shape)
```

```python
xbar = np.mean(Xsub, axis=1)
xbar = xbar.reshape(4096, 1)
```

```python
print(xbar.shape)
```

```python
mean_face = xbar.reshape(64, 64)
```

```python
ID_last_two_digits = 45
```

```python
my_face = Xsub[:, ID_last_two_digits].reshape(64, 64)
```

```python
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(mean_face, cmap=plt.cm.gray, vmin=0, vmax=1)
plt.title("Mean Face")
plt.axis('off')

# Plot the specific face image
plt.subplot(1, 2, 2)
plt.imshow(my_face, cmap=plt.cm.gray, vmin=0, vmax=1)
plt.title(f"Face {ID_last_two_digits}")
plt.axis('off')

plt.show()
```
