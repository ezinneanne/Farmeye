```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
A = np.array([[0.05,0.93,0.02,0.0],[0.10,0.86,0.04,0.0],[0.0,0.0,0.80,0.20],[0.50,0.10,0.0,0.4]])
```

```python
# Number of hours to simulate
time_steps = 24

# Initialize list to store the expected number of machines in each state at each hour
expected_machines = [initial_state]  # Initial state with all machines Working

# Loop over each hour to calculate expected machine distribution
for t in range(1, time_steps + 1):
    next_state = expected_machines[-1] @ A  # Compute next state distribution
    expected_machines.append(next_state)     # Store the result

# Convert list to numpy array for easier manipulation and display
expected_machines = np.array(expected_machines)
expected_machines
```

```python
initial_state = np.array([0, 200, 0, 0])  # [Idle, Working, Broken, Repair]
```

```python
# Plot the results
plt.figure(figsize=(10, 6))
plt.plot(expected_machines[:, 0], label="Idle")
plt.plot(expected_machines[:, 1], label="Working")
plt.plot(expected_machines[:, 2], label="Broken")
plt.plot(expected_machines[:, 3], label="Repair")

# Add titles and labels
plt.title("Machine State Distribution Over Time")
plt.xlabel("Time (hours)")
plt.ylabel("Number of Machines")
plt.legend(loc="upper right")
plt.grid()
plt.show()
```

```python
# Calculate the eigenvalues and eigenvectors of A
eigenvalues, eigenvectors = np.linalg.eig(A)

# Display eigenvalues and eigenvectors
print("Eigenvalues:", eigenvalues)
print("Eigenvectors:\n", eigenvectors)
```

```python
# Initial state: all machines are Working
initial_state = np.array([0, 200, 0, 0, 0])

# Track time steps and states
time_steps = 0
current_state = initial_state

# Loop until Working machines drop below 20
while current_state[1] >= 20:
    current_state = current_state @ A  # Update state distribution
    time_steps += 1

# Results
time_steps, current_state
# for answer 3a
```
