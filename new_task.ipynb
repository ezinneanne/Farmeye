```python
def calculate_efficiency(a, b, allocation):
    n = len(allocation)
    primary_time = [0] * n
    secondary_time = [0] * n

    for i in range(n):
        idx = allocation[i]
        if i == 0:
            primary_time[i] = a[idx]
            secondary_time[i] = primary_time[i] + b[idx]
        else:
            primary_time[i] = primary_time[i - 1] + a[idx]
            secondary_time[i] = max(secondary_time[i - 1], primary_time[i]) + b[idx]

    return primary_time, secondary_time


# Inputs
a = [3, 2, 4, 1, 5]  # Primary processing times
b = [2, 3, 1, 4, 2]  # Secondary processing times
allocation = [1, 0, 4, 3, 2]

# Calculate Efficiency
primary_time, secondary_time = calculate_efficiency(a, b, allocation)

# Output Results
print("Primary Time:", primary_time)
print("Secondary Time:", secondary_time)
```

```python
import random

def roll_die():
    """Simulates rolling a 6-sided die."""
    return random.randint(1, 6)

def free_throw(shooter, defender):
    """
    Simulates a single free throw in the HoopShot game.
    
    Args:
        shooter (str): Name of the shooter player.
        defender (str): Name of the defender player.
    
    Returns:
        tuple: (point_winner, description)
            point_winner (str): Name of the player who won the point or None if unresolved.
            description (str): Narrative of the actions during the throw.
    """
    narration = []
    narration.append(f"{shooter} takes the shot!")
    
    # Shooter rolls the green die
    green_roll = roll_die()
    narration.append(f"{shooter} rolls the green die and gets {green_roll}.")
    
    if green_roll in [1, 2]:
        narration.append("It's a clean basket! Shooter scores!")
        return shooter, "\n".join(narration)
    elif green_roll == 3:
        narration.append("The shot is a miss! Defender scores!")
        return defender, "\n".join(narration)
    else:
        narration.append("The shot is in progress...")

    # Defender rolls the orange die
    orange_roll = roll_die()
    narration.append(f"{defender} rolls the orange die and gets {orange_roll}.")
    
    if orange_roll == 1:
        narration.append("It's a block! Defender scores!")
        return defender, "\n".join(narration)
    elif orange_roll == 2:
        narration.append("Foul! Shooter gets another free throw.")
        return free_throw(shooter, defender)  # Recursive call for a new free throw
    else:
        narration.append("Pressure applied! The shooter rolls again.")

    # Shooter rolls the purple die
    purple_roll = roll_die()
    narration.append(f"{shooter} rolls the purple die and gets {purple_roll}.")
    
    if purple_roll == 1:
        narration.append("Basket made! Shooter scores!")
        return shooter, "\n".join(narration)
    elif purple_roll == 2:
        narration.append("Out of bounds! Defender scores!")
        return defender, "\n".join(narration)
    else:
        narration.append("Rebound fight! The point remains unresolved.")
        return None, "\n".join(narration)

def play_game(player1, player2):
    """
    Simulates a complete HoopShot game between two players.
    
    Args:
        player1 (str): Name of player 1.
        player2 (str): Name of player 2.
    
    Returns:
        None
    """
    scores = {player1: 0, player2: 0}
    print(f"Game start! {player1} vs. {player2}\n")
    
    while scores[player1] < 5 and scores[player2] < 5:
        print(f"Current score: {player1} {scores[player1]} - {player2} {scores[player2]}")
        shooter, defender = (player1, player2) if scores[player1] <= scores[player2] else (player2, player1)
        print(f"\n{shooter} steps up for the free throw!")
        
        point_winner, description = free_throw(shooter, defender)
        print(description)
        
        if point_winner:
            scores[point_winner] += 1
            print(f"{point_winner} scores a point!")

    winner = player1 if scores[player1] == 5 else player2
    print(f"\nFinal score: {player1} {scores[player1]} - {player2} {scores[player2]}")
    print(f"Game over! {winner} wins!\n")

# Example game simulation
play_game("Alice", "Bob")
```

```python
import random

def roll_die():
    """Simulates rolling a 6-sided die."""
    return random.randint(1, 6)

def simulate_free_throw(shooter, defender, stats):
    """
    Simulates a single free throw sequence in HoopShot.

    Args:
        shooter (str): Name of the shooter.
        defender (str): Name of the defender.
        stats (dict): Dictionary to track game statistics.

    Returns:
        tuple: (winner, narrative)
    """
    narrative = []
    narrative.append(f"{shooter} steps up for the free throw...")

    # Shooter rolls the green die
    green_roll = roll_die()
    narrative.append(f"{shooter} rolls the green die: {green_roll}.")
    if green_roll in [1, 2]:
        narrative.append("It's a clean basket! Shooter scores!")
        return shooter, "\n".join(narrative)
    elif green_roll == 3:
        narrative.append("Missed shot! Defender wins the point!")
        return defender, "\n".join(narrative)
    else:
        narrative.append("The shot is in progress...")

    # Defender rolls the orange die
    orange_roll = roll_die()
    narrative.append(f"{defender} rolls the orange die: {orange_roll}.")
    if orange_roll == 1:
        stats['blocks'] += 1
        narrative.append("Blocked by the defender! Defender wins the point!")
        return defender, "\n".join(narrative)
    elif orange_roll == 2:
        stats['fouls'] += 1
        narrative.append("Foul by the defender! Shooter gets another chance!")
        return simulate_free_throw(shooter, defender, stats)  # Recursive call
    else:
        narrative.append("Pressure applied! The shooter must roll again.")

    # Shooter rolls the purple die
    purple_roll = roll_die()
    narrative.append(f"{shooter} rolls the purple die: {purple_roll}.")
    if purple_roll == 1:
        narrative.append("Basket made! Shooter scores!")
        return shooter, "\n".join(narrative)
    elif purple_roll == 2:
        narrative.append("Out of bounds! Defender wins the point!")
        return defender, "\n".join(narrative)
    else:
        stats['rebounds'] += 1
        narrative.append("Rebound fight! The point remains unresolved.")
        return None, "\n".join(narrative)

def simulate_game(player1, player2, skew_shooter=False, start_scores=(0, 0), simulations=1):
    """
    Simulates a full HoopShot game.

    Args:
        player1 (str): Name of player 1.
        player2 (str): Name of player 2.
        skew_shooter (bool): If True, shooter success is biased.
        start_scores (tuple): Starting scores (player1, player2).
        simulations (int): Number of games to simulate for probability calculation.

    Returns:
        dict: Game statistics and results.
    """
    total_stats = {'shooter_wins': 0, 'blocks': 0, 'fouls': 0, 'rebounds': 0}

    for _ in range(simulations):
        scores = {player1: start_scores[0], player2: start_scores[1]}
        stats = {'blocks': 0, 'fouls': 0, 'rebounds': 0}

        while scores[player1] < 5 and scores[player2] < 5:
            shooter, defender = (player1, player2) if random.random() < (0.7 if skew_shooter else 0.5) else (player2, player1)
            winner, _ = simulate_free_throw(shooter, defender, stats)
            if winner:
                scores[winner] += 1

        if scores[player1] == 5:
            total_stats['shooter_wins'] += 1
        total_stats['blocks'] += stats['blocks']
        total_stats['fouls'] += stats['fouls']
        total_stats['rebounds'] += stats['rebounds']

    return total_stats

# Run Simulations
def analyze_game():
    player1, player2 = "Shooter", "Defender"
    print("Simulating normal games...")
    normal_stats = simulate_game(player1, player2, simulations=10000)
    print(f"Shooter win rate: {normal_stats['shooter_wins'] / 10000:.2%}")
    print(f"Average blocks per game: {normal_stats['blocks'] / 10000:.2f}")
    print(f"Average fouls per game: {normal_stats['fouls'] / 10000:.2f}")
    print(f"Average rebounds per game: {normal_stats['rebounds'] / 10000:.2f}\n")

    print("Simulating games with a 3-1 lead for the defender...")
    defender_lead_stats = simulate_game(player1, player2, start_scores=(1, 3), simulations=10000)
    print(f"Shooter win rate: {defender_lead_stats['shooter_wins'] / 10000:.2%}")
    print(f"Average blocks per game: {defender_lead_stats['blocks'] / 10000:.2f}")
    print(f"Average fouls per game: {defender_lead_stats['fouls'] / 10000:.2f}")
    print(f"Average rebounds per game: {defender_lead_stats['rebounds'] / 10000:.2f}\n")

    print("Simulating games with a 3-1 lead for the shooter...")
    shooter_lead_stats = simulate_game(player1, player2, start_scores=(3, 1), simulations=10000)
    print(f"Shooter win rate: {shooter_lead_stats['shooter_wins'] / 10000:.2%}")
    print(f"Average blocks per game: {shooter_lead_stats['blocks'] / 10000:.2f}")
    print(f"Average fouls per game: {shooter_lead_stats['fouls'] / 10000:.2f}")
    print(f"Average rebounds per game: {shooter_lead_stats['rebounds'] / 10000:.2f}\n")

# Analyze the game
analyze_game()
```
