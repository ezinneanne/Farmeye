```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('coviduk.csv')
```

```python
print(data.head())
```

```python
print(data.columns)
```

```python
data['date'] = pd.to_datetime(data['date'])
```

```python
plt.figure(figsize=(10, 6))

# Create the line plot
sns.lineplot(data=data, x='date', y='metric_value', label='Metric Value', color='b')

# Fill the area under the line
plt.fill_between(data['date'], data['metric_value'], color='blue', alpha=0.3)

# Add title and labels
plt.title('COVID-19 Healthcare Admissions Over Time', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Metric Value', fontsize=12)

# Rotate x-axis ticks for readability
plt.xticks(rotation=45)

# Add legend
plt.legend()

# Adjust layout for better fit
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import plotly.express as px
```

```python
fig = px.line(data, x='date', y='metric_value', title='COVID-19 Healthcare Admissions Over Time',
              labels={'date': 'Date', 'metric_value': 'Metric Value'},
              markers=True)
```

```python
fig.update_layout(
    title_font=dict(size=18),
    xaxis_title="Date",
    yaxis_title="Metric Value",
    xaxis=dict(tickangle=45, dtick="M1"),  # Ensure monthly ticks are shown
    template="plotly_white"
)

# Add color fill to the line plot
fig.update_traces(
    line=dict(color='blue', width=2),
    fill='tozeroy'  # Fills the area below the line
)

# Show the interactive plot
fig.show()
```
